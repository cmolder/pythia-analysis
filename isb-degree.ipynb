## ISB real vs. ISB ideal

```python
from importlib import reload
reload(plots)

isb_df = {
    'ISB Real (depth 1)': dfs['baseline']['isb_real_1'],
    'ISB Real (depth 2)': dfs['baseline']['isb_real_2'],
    'ISB Ideal (depth 1)': dfs['baseline']['isb_ideal_1'],
    'ISB Ideal (depth 2)': dfs['baseline']['isb_ideal_2'],
}

real_cmap = cm.get_cmap('Oranges')
ideal_cmap = cm.get_cmap('Blues')
real_cmap_idxs = np.linspace(0.5, 0.8, 2)
ideal_cmap_idxs = np.linspace(0.5, 0.8, 2)

isb_colors = {
    'ISB Real (depth 1)': real_cmap(real_cmap_idxs[0]),
    'ISB Real (depth 2)': real_cmap(real_cmap_idxs[1]),
    'ISB Ideal (depth 1)': ideal_cmap(ideal_cmap_idxs[0]),
    'ISB Ideal (depth 2)': ideal_cmap(ideal_cmap_idxs[1]),
}
legend_kwargs = {
    'fontsize': 8,
}
plots.plot_everything(
    isb_df, suites=suites, metrics=metrics, colors=isb_colors, 
    legend_kwargs=legend_kwargs, figsize=(9, 5))
```
