```python
import numpy as np
import pandas as pd
from scipy import stats
from IPython.display import display

from utils import tables
from utils import utils
```

```python
cygnus_stateless = '/u/cmolder/GitHub/ChampSim/out/pythia/degree1_featureless_tuned_nozeroaction.csv'
cygnus_pc = '/u/cmolder/GitHub/ChampSim/out/pythia/degree1_tuned_nozeroaction.csv'
weights = '/scratch/cluster/cmolder/traces/spec06_gap/champsim/weights.txt'

cygnus_stateless = tables.load_stats_csv(cygnus_stateless, ['cygnus'], prefetchers_level='l2')['cygnus']
cygnus_pc = tables.load_stats_csv(cygnus_pc, ['cygnus_state'], prefetchers_level='l2')['cygnus_state']

phase_weights = {}
with open(weights) as f:
    for line in f:
        phase_weights[line.split()[0]] = float(line.split()[1])
```

```python
pc_ipcs = []
stateless_ipcs = []
best_ipcs = []

for trace in utils.suites['spec06']:
    cygnus_pc_tr = np.array(cygnus_pc[(cygnus_pc.cpu0_trace == trace) & (cygnus_pc.cpu0_simpoint != 'weighted')].ipc_improvement)
    cygnus_stateless_tr = np.array(cygnus_stateless[(cygnus_stateless.cpu0_trace == trace) & (cygnus_stateless.cpu0_simpoint != 'weighted')].ipc_improvement)
    phases = np.array(cygnus_pc[(cygnus_pc.cpu0_trace == trace) & (cygnus_pc.cpu0_simpoint != 'weighted')].cpu0_simpoint)
    weights = np.array([phase_weights[f'{trace}_{phase}'] if phase != 'default' else phase_weights[trace] for phase in phases])

    cygnus_best_tr = np.maximum(cygnus_pc_tr, cygnus_stateless_tr)
    #display(cygnus_pc_tr)
    #display(cygnus_stateless_tr)
    # print(trace)
    # print(phases, weights)
    # print(cygnus_pc_tr)
    # print(cygnus_stateless_tr)
    # print(cygnus_best_tr)

    pc_ipcs.append(stats.gmean(100 + cygnus_pc_tr, weights=weights) - 100)
    stateless_ipcs.append(stats.gmean(100 + cygnus_stateless_tr, weights=weights) - 100)
    best_ipcs.append(stats.gmean(100 + cygnus_best_tr, weights=weights) - 100)
```

```python
print('PC IPC:', stats.gmean(100 + np.array(pc_ipcs)) - 100)
print('Stateless IPC:', stats.gmean(100 + np.array(stateless_ipcs)) - 100)
print('Best IPC:', stats.gmean(100 + np.array(best_ipcs)) - 100)
```
