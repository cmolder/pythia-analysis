# Bandwidth studies

```python
def cycles_to_ns(cycles, freqs):
    return cycles * (1000 / freqs)

def mtps_to_gibps(x):
    return x * 8 * 0.000931322574615478515625

def gibps_to_mtps(x):
    return x / 8 / 0.000931322574615478515625

def mtps_to_gbps(x):
    return x * 8 * 0.001

def gbps_to_mtps(x):
    return x / 8 / 0.001

def plot_dram_sweep(data: Dict[int, dict],
                    statistic_name: str = "cpu.gmean.ipc_improvement",
                    suites: Optional[List[str]] = None,
                    ymin: Optional[float] = None,
                    ymax: Optional[float] = None,
                    colors: Optional[dict] = defaultdict(lambda: None),
                    markers: Optional[dict] = defaultdict(lambda: "o"),
                    legend_kwargs: Optional[dict] = {},
                    ylabel: Optional[str] = None):

    fig, ax = plt.subplots(figsize=(5, 3), dpi=dpi)
    #secax = ax.secondary_xaxis('top', functions=(mtps_to_gibps, gibps_to_mtps))
    secax = ax.secondary_xaxis('top', functions=(mtps_to_gbps, gbps_to_mtps))

    # Generate tables for each bandwidth
    tables = {}
    for bw, experiments in data.items():
        tables[bw] = plot.get_suite_table(experiments, statistic_name,
                                          suites=suites, add_mean=True)
    # Collate tables
    exp_ys = defaultdict(lambda: [])
    for bw, table in tables.items():
        for exp, row in table.iterrows():
            exp_ys[exp].append(row.iloc[-1]) # gmean/amean

    # Plot
    for exp, ys in exp_ys.items():
        Xs = np.array(list(data.keys()), dtype=np.float32)
        ys = np.array(ys, dtype=np.float32)
        ax.plot(Xs, ys, label=exp, 
                color=colors[exp], 
                marker=markers[exp], markersize=5)

    # X = np.array(list(csim_data_tr.keys()), dtype=np.float32)
    # y = np.array(list(csim_data_tr.values()), dtype=np.float32)

    # ax.plot(X, y,
    #             label="ChampSim" if i == 0 else None, 
    #             marker='o', markersize=3, 
    #             color='C0')

    ax.grid(axis='both', color='lightgray')
    # ax.set_ylim(ymin=ymin, ymax=ymax)
    ax.set_xlim(xmin=0)

    ax.xaxis.set_tick_params(labelsize=7.5)
    ax.yaxis.set_tick_params(labelsize=7.5)
    secax.xaxis.set_tick_params(labelsize=7.5)

    ax.set_xlabel("DRAM data rate (MT/s)", fontsize=7.5)
    #secax.set_xlabel('DRAM bandwidth (GiB/s)', fontsize=7.5)
    secax.set_xlabel("DRAM bandwidth (GB/s)", fontsize=7.5)
    ax.set_ylabel(ylabel, fontsize=7.5)
    
    fig.legend(**legend_kwargs)
    fig.tight_layout()

def table_dram_sweep_suite(data: Dict[int, dict],
                           statistic_name: str = "cpu.gmean.ipc_improvement",
                           suites: Optional[List[str]] = None):
    for bw, dat in data.items():
        table = plot.get_suite_table(dat, statistic_name,
                                     suites=suites, add_mean=True).T
        print(f"=== {bw} MT/s ===")
        display(table)

def table_dram_sweep_benchmark(data: Dict[int, dict],
                               suite: str,
                               statistic_name: str = "cpu.gmean.ipc_improvement"):
    for bw, dat in data.items():
        table = plot.get_benchmark_table(dat, suite, statistic_name).T
        print(f"=== {bw} MT/s ===")
        display(table)
```

```python
degreedyn_1core_dfs = {}
for bw in bandwidths:
    degreedyn_1core_dfs[bw] = {
        "Cygnus" : studies[f"prefetchers__{bw}"]["cygnus"],
        "my5"    : studies[f"prefetchers__{bw}"]["my5"],
        "Pythia" : studies[f"prefetchers__{bw}"]["pythia"],
        "Bandit" : studies[f"prefetchers__{bw}"]["mab"],
        "SPP+PPF": studies[f"prefetchers__{bw}"]["spp_ppf"],
        #"SPP"    : studies[f"prefetchers__{bw}"]["spp"],
        #"Bingo"  : studies[f"prefetchers__{bw}"]["bingo"],
    }

degreedyn_1core_colors = defaultdict(lambda: None, {
    "Cygnus" : colormaps["Set3"](0),
    "my5"    : colormaps["Set3"](9),
    "Pythia" : colormaps["Set3"](2),
    "Bandit" : colormaps["Set3"](5),
    "Bingo"  : colormaps["Set3"](3),
    #"SMS": colormaps["Set3"](4),
    "SPP"    : colormaps["Set3"](5),
    "SPP+PPF": colormaps["Set3"](6),
    #"MLOP": colormaps["Set3"](7),
    #"DSPatch": colormaps["Set3"](8),
    #"BO-Width": colormaps["Set3"](9),
    #"Cygnus-Depth": colormaps["Set3"](1),
})
degreedyn_1core_markers = defaultdict(lambda: "o", {
    "Cygnus" : "o",
    "my5"    : "*",
    "Pythia" : "^",
    "SPP+PPF": "v",
    "Bandit" : "|",
})
degreedyn_1core_legend_kwargs = {
    "ncol": 6,
    "fontsize": 6.75,
    "loc": "lower center",
    "bbox_to_anchor": (0.5, 1.0),
    "framealpha": 0,
}
```
